```python
import requests
from bs4 import BeautifulSoup
import re

url = "http://books.toscrape.com/"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

books = soup.select("article.product_pod")

for book in books:
    title = book.h3.a["title"]
    price_str = book.select_one("p.price_color").text
    price_clean = re.sub(r'[^\d\.]', '', price_str)
    price = float(price_clean)

    # Rating is in class like "star-rating Three"
    rating_class = book.select_one("p.star-rating")["class"]
    rating_words = ['Zero', 'One', 'Two', 'Three', 'Four', 'Five']
    rating = next((i for i, word in enumerate(rating_words) if word in rating_class), 0)

    print(f"Title: {title}")
    print(f"Price: £{price}")
    print(f"Rating: {rating} stars")
    print("-" * 30)
```

```python
import requests
from bs4 import BeautifulSoup
import re
import time

all_books = []

for page_num in range(1, 6):  # first 5 pages
    if page_num == 1:
        url = "http://books.toscrape.com/"
    else:
        url = f"http://books.toscrape.com/catalogue/page-{page_num}.html"

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    books = soup.select("article.product_pod")

    for book in books:
        title = book.h3.a["title"]
        price_str = book.select_one("p.price_color").text
        price_clean = re.sub(r'[^\d\.]', '', price_str)
        price = float(price_clean)

        rating_class = book.select_one("p.star-rating")["class"]
        rating_words = ['Zero', 'One', 'Two', 'Three', 'Four', 'Five']
        rating = next((i for i, word in enumerate(rating_words) if word in rating_class), 0)

        all_books.append({
            "title": title,
            "price": price,
            "rating": rating
        })

    time.sleep(1)  # be polite, avoid hammering server too fast

print(f"Total books scraped: {len(all_books)}")

# Just print first 5 books to check
for book in all_books[:5]:
    print(book)
```

```python
import sqlite3

# Connect (or create) a database file
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

# Create a table for the books
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    price REAL,
    rating INTEGER
)
''')

# Insert books into the table
for book in all_books:
    cursor.execute('''
    INSERT INTO books (title, price, rating)
    VALUES (?, ?, ?)
    ''', (book['title'], book['price'], book['rating']))

conn.commit()
print("Books saved to database successfully!")

# Close connection
conn.close()
```

```python
books_data = []

for book in books:
    # Title
    title = book.h3.a['title']

    # Price (remove £ and convert to float)
    price_str = book.select_one("p.price_color").text
    price = float(price_str.replace('£', '').strip())

    # Rating (from class attribute, e.g. 'star-rating Three')
    rating_class = book.find("p", class_="star-rating")["class"]
    # The rating is the second class (e.g. 'Three')
    rating_text = [c for c in rating_class if c != "star-rating"][0]
    # Convert rating text to number
    rating_dict = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
    rating = rating_dict.get(rating_text, 0)

    # Append dictionary
    books_data.append({
        'title': title,
        'price': price,
        'rating': rating
    })

# Now create DataFrame
import pandas as pd
df = pd.DataFrame(books_data)
print(df.head())
```

```python
# Save DataFrame to CSV
df.to_csv('books_data.csv', index=False)
print("Data saved to books_data.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

books = []

for page in range(1, 51):  # loop through all 50 pages
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    articles = soup.find_all('article', class_='product_pod')

    for article in articles:
        title = article.h3.a['title']
        price = float(article.find('p', class_='price_color').text[1:])  # remove '£'

        # Extract rating from class attribute
        rating_class = article.p['class'][1]
        rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
        rating = rating_map.get(rating_class, 0)

        books.append({
            'title': title,
            'price': price,
            'rating': rating
        })

# Convert to DataFrame
df = pd.DataFrame(books)

# Save to CSV
df.to_csv("books_data_all_pages.csv", index=False)

# Display first few rows
df.head()
```
